## Activity: Discover what is in your dataset

This activity uses a dataset called `Unicorn_Companies.csv`. It represents a list of private companies with a value of over 1 billion USD as of March 2022. The data includes the name of the country where the company was founded, its current valuation, funding, industry, top investors, year it was founded, and the year it reached a 1 billion USD valuation.

The dataset contains: 1,074 rows and 10 columns.
The metadata is given below.



```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
companies = pd.read_csv('./resources/Unicorn_Companies.csv')
companies.head(10)
```

```python
companies.info()
```

```python
companies.describe()
```

```python
# change date to datetime format
companies['Date Joined'] = pd.to_datetime(companies['Date Joined'])
```

```python
companies.info()
```

```python
# get the year from Year joined and as a new column
companies['Year Joined'] = companies['Date Joined'].dt.year
```

```python
companies.head()
```

```python
# sample from the data

companies_sampled = companies.sample(n=50, random_state=21)
companies_sampled.head()
```

```python
# visualize time (in years) it took to reach unicorn status

# data prep
companies_sampled['years_until_unicorn'] = companies_sampled['Year Joined'] - companies_sampled['Year Founded']
grouped = companies_sampled[['Industry', 'years_until_unicorn']].groupby("Industry").max()
grouped
```

```python
plt.bar(grouped.index, grouped['years_until_unicorn'])
plt.xticks(rotation=45, horizontalalignment='right')
plt.xlabel('Industry')
plt.ylabel('Max no. of years to reach unicorn status')

plt.show()
```
